```python
from matplotlib import pyplot as plt
from pylab import legend
ratio_of_FreiHAND = [0,4,7.7,14.3,20,25,30,33.3,36.8,40,42.9,50]
MPJPE = [4.71,3.03,2.77,2.92,2.82,3.07,2.80,4.84,2.75,2.86,2.99,2.91]
plt.plot(ratio_of_FreiHAND, MPJPE, marker = 'o')
plt.xlabel('ratio of FreiHAND')
plt.ylabel('MPJPE(mm)')
plt.title('MPJPE according to ratio of FreiHAND')
legend(['MPJPE'], loc='best')
plt.show()
```
